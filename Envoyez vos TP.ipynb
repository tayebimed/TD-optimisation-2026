```python
# ==========================
# 📌 Informations de l'étudiant
# ==========================

nom = "Idouaouzal"
prenom = "nora"
groupe = "3"

# Fichier du TP à envoyer
file = "correction_exercice_1.ipynb"
```

```python
import smtplib
from email.message import EmailMessage
import os
from dotenv import load_dotenv
from datetime import datetime

date_envoi = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
# Charger .env
load_dotenv()

email_sender = os.getenv("EMAIL_SENDER")
email_password = os.getenv("EMAIL_PASSWORD")
email_receiver = os.getenv("EMAIL_RECEIVER")

msg = EmailMessage()
msg['Subject'] = "TP automatique"
msg['From'] = email_sender
msg['To'] = email_receiver
msg.set_content(
    f"Bonjour Monsieur,\n\n"
    f"Veuillez trouver ci-joint mon Travail Pratique (TP).\n\n"
    f"Informations de l'étudiant :\n"
    f"Nom : {nom}\n"
    f"Prénom : {prenom}\n"
    f"Groupe : {groupe}\n\n"
    f"Date d’envoi : {date_envoi}\n\n"
    f"Cordialement."
)

with open(file, 'rb') as f:
    msg.add_attachment(f.read(), maintype='application', subtype='octet-stream', filename=file)

with smtplib.SMTP_SSL("smtp.gmail.com", 465) as smtp:
    smtp.login(email_sender, email_password)
    smtp.send_message(msg)

print("Email envoyé !")
```
